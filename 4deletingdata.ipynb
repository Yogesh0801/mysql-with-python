```python
import mysql.connector
mydb = mysql.connector.connect(
  host="localhost",
    user="root",
    password="root",
    database="mydatabase"
)
mycursor = mydb.cursor()
mycursor.execute("delete from students where phone=78323")
mydb.commit()
print(mycursor.rowcount, "record(s) deleted")
```
